```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('/content/train_mfcc_data.csv')
```

```python
train['age'].value_counts(ascending=True).unique()
```

```python
bins = list(range(0, 133, 10))
```

```python
bins_label = [str(x)+"s" for x in bins]
```

```python
bins_label
```

```python
train["age"] = pd.cut(train["age"], bins, right=False, labels=bins_label[:-1])
```

```python
train
```

```python
train.groupby(['age', 'covid19']).count()
```

```python
train_negative = train[train['covid19'] ==0]
```

```python
train_negative
```

```python
train_negative.mean()
```

```python
train_positive = train[train['covid19'] ==1]
```
